```python
#Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#import terrorism dataset
terror_df=pd.read_csv("C:\\Users\\pjanh\\Desktop\\globalterrorismdb_0718dist.csv",encoding='ISO-8859-1')
```

```python
print("Number of rows in the dataset : {}".format(terror_df.shape[0]))
print("Number of columns in the dataset : {}".format(terror_df.shape[1]))
```

```python
#Renaming the column names 
terror_df=terror_df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region',
                             'provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type','targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
```

```python
#Creating a new dataset
terror_df=terror_df[['Year','Month','Day','Country','Region','State','City','Attack_type','Target_type','Target_subtype','Gang_name','Weapon_type']]
```

```python
#Display the dataset
terror_df
```

```python
#Display number of missing values 
terror_df.isnull().sum()
```

```python
x_year = terror_df['Year'].unique()
y_count_years = terror_df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = x_year,
           y = y_count_years,
           palette = 'mako')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=terror_df,palette='flare')
plt.xticks(rotation=45)
plt.title('Number Of Terrorist Activities Each Year')
plt.show()
```

```python
pd.crosstab(terror_df.Year, terror_df.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

```python
terror_df.Country.value_counts()[:15]
```

```python
#Top Countries affected by Terror Attacks¶
mostGlob = terror_df.Country.value_counts()[:21]
plt.figure(figsize=(16,12))
ax = sns.barplot(x=mostGlob,y=mostGlob.index,palette="rocket")
plt.title("Most 20 Countries Attacks")
plt.xlabel("Count")
plt.show()
```

```python
city=terror_df.City.value_counts()
city=city[1:]
plt.figure(figsize=(13,7))
city=city.head(20)
sns.barplot(city.index,city,palette='cubehelix').set_title('Top 20  most affected cities')
plt.xticks(rotation=90)
plt.xlabel("Cities")
plt.ylabel("Count")
plt.show()
```
